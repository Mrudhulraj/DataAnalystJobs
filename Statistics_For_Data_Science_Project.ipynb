```python
df["Headquarters"].value_counts().describe() 
mean=df["Headquarters"].value_counts().mean() 
median=df["Headquarters"].value_counts().median() 
mode=df["Headquarters"].mode() 

print("Mean is:",mean)  
print("Median is:",median)
print("Mode is:",mode) 
```

As can be seen, Mean>Median and also by looking at the above boxplot and histogram, it is clear that the above graph is Right Skewed or Positively Skewed 

Thus Mode is the highest value, which is New York, Thus one must try to apply in New York as it has more headquaters, resulting in more job oppurtunities.
